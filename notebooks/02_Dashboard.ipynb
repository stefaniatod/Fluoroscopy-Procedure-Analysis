```python
import pandas as pd
```

```python
import plotly.express as px
```

```python
df=pd.read_csv("../data/cleaned_data.csv")
```

```python
df.head()
```

```python
fig = px.bar(df.groupby("Procedure", as_index=False)["DAP_Gy_cm2"].mean(),
             x="Procedure",
             y="DAP_Gy_cm2",
             title="Average DAP by Procedure")
fig.show()
fig.write_html("../images/average_dap_gy_cm2_by_procedure.html")
```

```python
fig=px.bar(df.groupby("Procedure", as_index=False)["Fluoroscopy_Time"].mean(),
           x="Procedure",
           y="Fluoroscopy_Time",
           title="Average Fluoroscopy Time by Procedure")
fig.show()
fig.write_html("../images/fluoroscopy_time.html")
```

```python
fig=px.histogram(df,
                 x="Patient_Age",
                 title="Patient Age Distribution")
fig.show()
fig.write_html("../images/patient_age_distribution.html")
```
